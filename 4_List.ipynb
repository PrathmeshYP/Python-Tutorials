```python
fruits=["Apple","Banana","Berry","Cherry","Pineapple","Chicku"]
print(fruits)
```

# List Methods

```python
print(len(fruits))
```

```python
fruits.append("Mango")
print(fruits)
```

```python
fruits.insert(4,"Strawberry")
print(fruits)
```

```python
fruits.remove("Apple")
print(fruits)
```

```python
fruits.pop()
print(fruits)
```

```python
fruits.sort()
print(fruits)
```

```python
fruits.reverse()
print(fruits)
```

```python
fruits.clear()
print(fruits)
```
