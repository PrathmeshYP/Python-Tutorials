# String Methods

```python
#Upper case

word="Pratham"
print(word.upper())
```

```python
#Lower case

word="Pratham"
print(word.lower())
```

```python
#First letter capital

word="pratham"
print(word.capitalize())
```

```python
#All words first letter capital

word="Pratham is smart boy"
print(word.title())
```

```python
#Replace

fruit="apple"
print(fruit.replace("a","A"))
```

```python
#split

print(fruit.split())
```

```python
#Find

sen ="hello, i am learning python"
print(sen.find("python"))
```

```python
# String Method strip
sent3 = "           Hello Wrold"
print(sent3.strip())   #Removes starting spaces
```
