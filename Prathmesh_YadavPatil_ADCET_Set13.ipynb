```python
# Select only Name, Department, and Final_Result columns.
df[["Name","Department","Final_Result"]]
```

```python
# Filter students who scored more than 80 in External Marks.
df[df["External_Marks"]>80][["Name","External_Marks"]]
```

```python
# Remove duplicate records.
df=df.drop_duplicates()
```

```python
# Fill missing values in Attendance column with mean.
df['Attendance_Percentage']=df["Attendance_Percentage"].fillna(df["Attendance_Percentage"].mean())
df
```

```python
# Replace missing marks with median value.
df['External_Marks']=df["External_Marks"].fillna(df["External_Marks"].median())
```

```python
df.info()
```

```python
# Standardize Gender column (convert all variations to Male/Female).
df["Gender"]=df["Gender"].replace({"F":"Female","M":"Male"})
```

```python
# Convert Age column into integer type.
df["Age"]=pd.to_numeric(df["Age"],errors="coerce")
```

```python
df.info()
```

```python
# Rename column names into lowercase.
df.columns=df.columns.str.lower()
df
```

```python
# Replace attendance values greater than 100 with 100.
df["attendance_percentage"]=pd.to_numeric(df["attendance_percentage"],errors="coerce")
df.info()
```

```python
df.loc[df['attendance_percentage'] > 100, 'attendance_percentage'] = 100 
```

```python
for i in df["attendance_percentage"]:
    if i > 100:
        print(i)
else:
    print("no value is greater than 100")
    
```

```python
# Change Department names to uppercase.
df.department=df.department.str.upper()
df
```

```python
#Find top 5 students based on Total Marks.
df["total_marks"]=df["internal_marks"]+df["external_marks"]
top_5_df = df.nlargest(5, columns='total_marks')
print(top_5_df)
```

```python
# Find students who failed.
df[df["total_marks"] < 80][["name", "total_marks"]]
```

```python
#Find highest attendance

higest_attendence=df.nlargest(1, columns='attendance_percentage')
print(higest_attendence)
```

```python
# Find student with minimum External Marks.
df.loc[df["external_marks"] == df["external_marks"].min(), ["name", "external_marks"]]
```

```python
# Create a new column “Percentage”
df["Percentage"]=(df["total_marks"]/200)*100
df
```

```python
# Use value_counts() on Final_Result column.
df["final_result"].value_counts()
```

```python
# Apply function to assign Grade based on Percentage.
def assign_grade(percentage):
    if percentage>=90:
        return 'A'
    elif percentage>=80:
        return 'B'
    elif percentage>=70:
        return 'C'
    elif percentage>=60:
        return 'D'
    else:
        return 'E'
df['Grade']=df['Percentage'].apply(assign_grade)
print(df)
```
