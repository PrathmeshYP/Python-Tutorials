```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("students (2).csv")
df
```

```python
sns.lineplot(data=df,x="Semester",y="Internal_Marks")
plt.show()
```

```python
sns.scatterplot(data=df,x='Internal_Marks',y='External_Marks',hue='Gender')
```

```python


sns.reset_defaults()
```

```python
sns.barplot(data=df,x='Department',y='Internal_Marks',hue='Gender')
plt.show()
```

```python
sns.countplot(data=df,x='Department')
plt.show()
```

```python
sns.histplot(data=df,x='Internal_Marks',bins=10,kde=True) #kernel Density estimation
plt.show()
```

```python
sns.kdeplot(data=df,x='Internal_Marks',fill=True)
```

```python
sns.boxplot(data=df,x='Gender',y='Internal_Marks',hue='Gender')
plt.show()
```

```python
sns.violinplot(data=df,x='Gender',y='Internal_Marks',hue='Gender')
plt.show()
```

```python
sns.stripplot(data=df,x='Gender',y='Internal_Marks')
plt.show()
```

```python
sns.swarmplot(data=df,x='Gender',y='Internal_Marks')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr)
plt.show()
```
