```python
marks = int(input("Enter marks:"))
if marks >= 85:
    print("Passed with grade A")
elif marks >= 65:
    print("Passed with grade B")
elif marks >= 55:
    print("Passed with grade C")
elif marks >= 35:
    print("Passed with grade D")
else:
    print("Failed!!")
```
