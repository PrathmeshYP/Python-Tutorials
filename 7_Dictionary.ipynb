# Dictionary

```python
#Dictionary declaration

stud={"name":"Prathmesh","age":22,"marks":{90,91,93},"College":"ADCET","name":"Prem"}
print(stud)
```

```python
print(stud["name"])
```

```python
#get method

print(stud.get("marks"))
```

```python
Employee={"Employee1":{"id":1,"name":"Pratham","Dept":"Data Analyst","salary":30000},                                              
          "Employee2":{"id":2,"name":"Prem","Dept":"Data Analyst","salary":20000},
          "Employee3":{"id":3,"name":"Prasad","Dept":"Data Scientist","salary":40000},
          "Employee4":{"id":4,"name":"Dhanraj","Dept":"Data Engineer","salary":35000},
          "Employee5":{"id":5,"name":"Saif","Dept":"Data Analyst","salary":25000}
         }
print(Employee)
```

```python
print(Employee.get("Employee1"))
print(Employee.get("Employee2"))
print(Employee.get("Employee3"))
print(Employee.get("Employee4"))
print(Employee.get("Employee5"))
```

```python
#keys method

print(Employee.keys())
```

```python
#Values method

print(Employee.values())
```

```python
#items method

print(Employee.items())
```

```python
#pop method

print(Employee.pop("Employee1"))
```

```python
print(Employee)
```

```python
#update method

print(Employee["Employee4"].update({"salary":60000}))
print(Employee)
```

```python
#clear method

print(Employee.clear())
print(Employee)
```
