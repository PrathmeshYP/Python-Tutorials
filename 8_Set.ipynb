# Set

```python
#set declaration

fruits={"Apple","Mango","Banana","Pineapple","Cherry"}
print(fruits)
```

```python
#add method

fruits.add("Chicku")
print(fruits)
```

```python
#update method

fruits.update({"Strawberry","Kiwi","Grapes"})
print(fruits)
```

```python
#remove method

fruits.remove("Kiwi")
print(fruits)
```

```python
#remove show error if the element is not present

fruits.remove("Kiwi")
print(fruits)
```

```python
#pop method

fruits.pop()
print(fruits)
```

```python
#discard method

fruits.discard("Grapes")
print(fruits)
```

```python
#discard doesnt show error if the element is not present

fruits.discard("Grapes")
print(fruits)
```

```python
#clear method

fruits.clear()
print(fruits)
```

```python
# union

a={10,20,30,40,45}
b={45,50,60,70,80}
a.union(b)
```

```python
# intersection

a.intersection(b)
```

```python
#difference

a.difference(b)
```

```python
b.difference(a)
```

```python
# symmetric_difference

a.symmetric_difference(b)
```

```python
# Isdisjoint
#Returns False if two awts have common element otherwise true

c={1,2}
d={1,2}
c.isdisjoint(d)
```

```python
# Issuperset
#Returns True if two sets have common elements otherwise false

c.issuperset(d)
```

```python
#Subset method
c.issubset(d)
```
