```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def area(self):
        return self.l * self.b
        
class Circle(Shape):
    def __init__(self,r):
        self.r=r
    def area(self):
        return 3.14 * self.r * self.r

r=Rectangle(10,5)
print("Area of rectangle is ",r.area())

c=Circle(5)
print("Area of circle is ",c.area())
    
```

```python
from abc import ABC,abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Bike(Vehicle):
    def start(self):
        print("Bike start with kick")

class Car(Vehicle):
    def start(self):
        print("car start with button")

b=Bike()
b.start()

c=Car()
c.start()
```
