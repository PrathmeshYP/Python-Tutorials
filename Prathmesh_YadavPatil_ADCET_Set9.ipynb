# 1) Create a base class Employee with method display()  and a derived  class manager that adds a method department.

```python
class Employee:
    def display(self):
        print("This is Employee")
class Manager(Employee):
    def department(self):
        print("Department is Finance")

m=Manager()
m.display()
m.department()
```

# 2) Write a program using multiple inheritance with classes account, loan and customer.

```python
class account:
    def show_1(self):
        print("This is credit account ")
class loan:
    def show_2(self):
        print("Youre elligible for loan")
class customer(account,loan):
    def show_3(self):
        print("This is customer class")

c=customer()
c.show_1()
c.show_2()
c.show_3()
        
```

# 3) Define a base class vehicle with a method start_engine(). Inherit it into car and add method drive().

```python
class vehicle:
    def start_engine(self):
        print("You can start engine")

class car(vehicle):
    def drive(self):
        print("You can drive")

c=car()
c.start_engine()
c.drive()
```

# 4) Create a class father and mother with methods skills() and hobby() respectively. Create a class child in that inherits from both and add its own method play ().

```python
class father:
    def skills1(self):
        print("Driving is skill of father")
    def hobby1(self):
        print("Reading books is my hobby of father")

class mother:
    def skills2(self):
        print("Cooking is skill of mother")
    def hobby2(self):
        print("Riding is my hobby of mother")

class child(father,mother):
    def play(self):
        print("Playing is my hobby of child")

c=child()
c.skills1()
c.hobby1()
c.skills2()
c.hobby2()
c.play()
```
