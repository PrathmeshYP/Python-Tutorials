# if else statement

```python
a=int(input("Enter a number"))
if a>10:
    print("Number is greater than 10")
else:
    print("Number is smaller than 10")
```

# Number is even or odd

```python
num=int(input("Enter a number : "))
if num%2==0:
    print("Number is even")
else:
    print("Number is odd")
```

# if-elif-else

```python
mark=int(input("Enter the mark:"))
if mark>=90:
    print("Greade A")
elif mark>=80:
    print("Greade B")
elif mark>=60:
    print("Greade C")
elif mark>=40:
    print("greade D")
else:
    print("fail")
```

# Positive or Negative

```python
num = int(input("Enter a number: "))

if num >= 0:
    if num % 2 == 0:
        print("Positive Even")
    else:
        print("Positive Odd")
else:
    print("Negative number")
```
