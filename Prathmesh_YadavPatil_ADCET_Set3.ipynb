# Write a program to check if the entered string is a palindrome or not.

```python
sen=input("Enter a string : ")
if sen == sen[::-1]:
    print("String is Palindrome")
else:
    print("String is not Palindrome")
```

# Take a character input and check whether it's a vowel or consonant.

```python
char=input("Enter a Character : ").lower()
if char in 'a,e,i,o,u':
    print("It is vowel")
else:
    print("It is consonant")
```

# Check if the entered string starts with a capital letter or not.

```python
sen=input("Enter any string : ")
if sen[0].isupper():
    print("String start with capital letter")
else:
    print("String start with lower letter")
```

# Check  the length of a word and categorize.
    a) if length <5 -> Short
    b) if length 5 to 8  -> Medium
    c) if 8 -> long

```python
word=input("Enter a word : ")
if len(word)<5:
    print("Short")
if len(word)>5 and len(word)<8:
    print("medium")
else:
    print("Long")
```
