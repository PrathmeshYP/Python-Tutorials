# 1)Write a program to check whether a person is a child, teenager, adult, or senior citizen based on age.

```python
age=int(input("Enter your age : "))
if age<14:
    print("Youre child")
elif age<18 and age>14:
    print("youre teenager")
elif age>18 and age<70:
    print("Your adult")
else:
    print("Youre senior citizen")
```

#  2)Take a number from the user and check whether it is divisible by both 3 and 5. 

```python
num = int(input("Enter a number : "))
if (num % 3==0) and (num % 5 == 0):
    print("Number is Divisible by 3 and 5")
else:
    print("Not divisible")
```

# 3)Input a number and print whether it is a single-digit, two-digit, or three-digit number.

```python
num=int(input("Enter a number"))
if num>0 and num<10:
    print("Single digit")
if num>9 and num<100:
    print("Two digit")
if num>99 and num<1000:
    print("Three digit")
else:
    print("Number is greater than three digit")
```

#  4)Take three sides of a triangle and check if the triangle is equilateral, isosceles, or scalene.

```python
a=int(input("Enter first side of triangle : "))
b=int(input("Enter second side of triangle : "))
c=int(input("Enter third side of triangle : "))

if a==b==c:
    print("Triangle is equilateral")
elif a==b or b==c or a==c:
    print("Triangle is isosceles")
else:
    print("Triangle is scalene")
```

# 6)Accept a temperature value and print: “Freezing” if temperature < 0 “Cold” if temperature is between 0 and 15 “Warm” if temperature is between 16 and 25 “Hot” if temperature > 25

```python
temp=int(input("Enter temperature : "))
if temp<0:
    print("Freezing")
elif temp>0 and temp<16:
    print("Cold")
elif temp>15 and temp<26:
    print("Warm")
else :
    print("Hot")
```

# 7)Take three numbers and check if their sum is even or odd. 

```python
a=int(input("Enter first number : "))
b=int(input("Enter second number : "))
c=int(input("Enter third number : "))
sum=a+b+c
if (sum%2==0):
    print("Sum is Even")
else:
    print("Sum is Odd")
```

# 8)Take marks of 5 subjects and print: “Excellent” if average >= 90 “Good” if average >= 75 “Average” if average >= 50 “Poor” if average < 50

```python
a=int(input("Enter first subject marks : "))
b=int(input("Enter second subject marks : "))
c=int(input("Enter third subject marks : "))
d=int(input("Enter fourth subject marks : "))
e=int(input("Enter fifth subject marks : "))
avg=(a+b+c+d+e)/5
if avg>=90:
    print("Excellent")
elif avg>=75:
    print("Good")
elif avg>=50:
    print("Average")
else:
    print("Poor")
```
