```python
df['Gender'].fillna("No Gender", inplace = True)
df.isnull().sum()
```

```python
mode=df['Senior Management'].mode().values[0]
df['Senior Management']=df['Senior Management'].fillna(mode)
df.isnull().sum()
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
sns.histplot(x='Salary', data=df,hue='Gender')
```
