```python
#Single Inheritance

class parent:
    def display(self):
        print("This is Parent class")
class child(parent):
    def show(self):
        print("This is child class")
c=child()
c.display()
c.show()
```

```python
#Multiple Inheritance

class Father:
    def skills(self):
        print("Driving")

class Mother:
    def talent(self):
        print("Cooking")

class child(Father,Mother):
    def display(self):
        print("This is child class and inherits from parents")

c=child()
c.skills()
c.talent()
c.display()
```

```python
class Employee:
    def __init__(self):
        self.company="Paarsh Infotech"

class Salary:
    def __init__(self):
        self.salary=80000

class Person(Employee,Salary):
    def __init__(self):
        Employee.__init__(self)
        Salary.__init__(self)

    def show(self):
        print(self.company)
        print(self.salary)
p=Person()
p.show()
```

```python
# Multilevel Inheritance

class A:
    def display_A(self):
        print("This is class A")
class B(A):
    def display_B(self):
        print("This is class B")
class C(B):
    def display_C(self):
        print("This is class C")
class D(C):
    def display_D(self):
        print("This is class D")

obj=D()
obj.display_A()
obj.display_B()
obj.display_C()
obj.display_D()
```

```python
class GrandParent:
    def display_grandparent(self):
        print("This is from Grandparent class")
class Parent(GrandParent):
    def display_parent(self):
        print("This is from Parent class")
class Child(Parent):
    def display_child(self):
        print("This is from Child class")

c=Child()
c.display_grandparent()
c.display_parent()
c.display_child()
```

```python
#hierarchical Inheritance
class Parent:
    def show_Parent(self):
        print("This is from parent class")
class Child1(Parent):
    def show_c1(self):
        print("This is from child 1 class")
class Child2(Parent):
    def show_c2(self):
        print("This is from child 2 class")

c1=Child1()
c1.show_Parent()
c1.show_c1()

c2=Child2()
c2.show_Parent()
c2.show_c2()
```

```python
class BankAccount:
    def account_type(self):
        print("This is a bank account")

class SavingAccount(BankAccount):
    def intreset(self):
        print("Intrest is 10%")

class CurrentAccount(BankAccount):
    def show(self):
        print("You can use youre current account")

s=SavingAccount()
s.account_type()
s.intreset()

c=CurrentAccount()
c.account_type()
c.show()
```

```python
class Person:
    def name(self):
        self.name="hybrid"
        print(self.name)

class Age(Person):
    def show(self):
        self.age=21
        print(self.age)
class height(Age):
    def display(self):
        self.height=5.8
        print(self.height)

a= height()
a.name()
a.show()
a.display()
```

```python
# Hybrid Inheritance

class A:
    def show_A(self):
        print("Class A method")

class B(A):
    def show_B(self):
        print("Class B method")

class C(A):
    def show_C(self):
        print("Class C method")

class D(B,C):
    def show_D(self):
        print("Class D method")

d=D()
d.show_A()
d.show_B()
d.show_C()
d.show_D()
```
