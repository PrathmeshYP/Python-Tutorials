# 6)Remove duplicates from a list.

```python
numbers=[1,2,3,2,3,4,2,5]
unique_lst=[]

for num in numbers:
    if num not in unique_lst:
        unique_lst.append(num)


print(unique_lst)
```

# 9) Sort list without using sort() function.

```python
numbers = [6,8,3,2,5,9]

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] > numbers[j]:
            numbers[i], numbers[j] = numbers[j], numbers[i]

print(numbers)
```

# 11) Merge two lists index-wise (L1[i] + L2[i]).

```python
L1=[2,3,4,5,6]
L2=[20,30,40,50,60]
merged=[]

for i in range(len(L1)):
    merged.append(L1[i] + L2[i])

print(merged)
```

# 12) Rotate a list (last element becomes first).

```python
numbers = [1, 2, 3, 4, 5]

last_element = numbers[-1]

rotated_list = [last_element]

for i in range(len(numbers) - 1):
    rotated_list.append(numbers[i])

print(rotated_list)
```

# 15) Convert list of lists into a single list (flatten list).

```python
nested_list = [[1, 2], [3, 4], [5, 6]]
flat_list = []

for sublist in nested_list:
    for item in sublist:
        flat_list.append(item)

print(flat_list)
```
