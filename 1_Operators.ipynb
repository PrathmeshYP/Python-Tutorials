# Logical Operators

```python
#and operator

age=int(input("Enter your age : "))
citizen= True
if age>=18 and citizen:
    print("Elligible for voting")
else:
    print("Not Elligible for voting")
```

```python
#or operator

age=int(input("Enter your age : "))
citizen= True
if age>=18 or citizen:
    print("Elligible for voting")
else:
    print("Not Elligible for voting")
```

```python
#not operator

x=True
y=False

print(not x)
print(not y)
```
