# 1)Create class Animal with method sound().
    Create child classes:
    Dog
    Cat
    Override sound() in both classes.

```python
class Animal:
    def sound(self):
        print("Animal Sound")

class Dog(Animal):
    def sound(self):
        super().sound()
        print("Dog barks")
        
class Cat(Animal):
    def sound(self):
        print("Cat mews")

d=Dog()
c=Cat()

d.sound()
c.sound()
```

# 2)Create class Shape with method area().
    Override it in:
    Rectangle
    Circle
    Print area for both objects.

```python

class Shape:
    def area(self):
        print("Area not defined")

class Rectangle(Shape):
    def area(self, length, width):
        print("Area of Rectangle =", length * width)

class Circle(Shape):
    def area(self, radius):
        pi=3.14
        print("Area of Circle =", pi * radius * radius)

r = Rectangle()
c = Circle()

r.area(5, 3)
c.area(4)

        
```

# 3)Create class Employee with method salary() returning 20000.
    Create:
    Manager (add bonus)
    Developer (add overtime pay)
    Override salary().

```python
class Employee:
    def salary(self):
        return 20000
class Manager(Employee):
    def salary(self):
        bonus=3000
        return super().salary()+bonus
class Developer(Employee):
    def salary(self):
        overtime=500
        return super().salary()+overtime

m=Manager()
d=Developer()
print("Manager Salary =", m.salary())
print("Developer Salary =", d.salary())
```

# 4)Create class Payment with method pay(amount).
    Override in:
    UPI
    CreditCard
    NetBanking
    Each prints different payment message.

```python
class Payment():
    def pay(self,amount):
        print("Payment amount = ",amount)

class UPI(Payment):
    def pay(self,amount):
        super().pay(amount)
        print("Payment through UPI = ",amount)

class Creditcard(Payment):
    def pay(self,amount):
        print("Payment through Credit card = ",amount)
    
class NetBanking(Payment):
    def pay(self,amount):
        print("Payment through Netbanking = ",amount)    

u=UPI()
c=Creditcard()
n=NetBanking()

u.pay(20000)
c.pay(30000)
n.pay(40000)
```

# 5)Create class Person with display().
    Child class Student should:
    call parent display using super()
    add roll number

```python
class Person:
    def display(self):
        print("Person")
class Student(Person):
    def display(self):
        super().display()
        self.roll_no = 101
        print("Roll Number =", self.roll_no)

s=Student()
s.display()
```
