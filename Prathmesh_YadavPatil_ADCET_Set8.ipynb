# Print the first 10 natural numbers using a loop.

```python
i=1
while i<=10:
    print(i)
    i+=1
    
```

# Print the multiplication table of a number entered by the user.

```python
a=int(input("Enter a number :"))
i=1
while i<=10:
    print(a," x ",i," = ",a*i)
    i+=1
```

# Write a program to calculate the sum of all numbers from 1 to n.

```python
n=int(input("Enter a number"))
i=1
sum=1
while i<=n:
    sum+=i
    print(sum)
    i+=1
    
```

# Create a list of 5 fruits and print each using a loop.

# Find the maximum and minimum number in a list of integers.

```python
nums = [4, 7, 1, 9, 2]

print("Maximum =", max(nums))
print("Minimum =", min(nums))
```

# Count how many times an element appears in a tuple.

```python
nums=(1,2,3,1,3,4,5,1,6)
count=0
a=int(input("Enter a number : "))
for i in nums:
    if i==a:
        count+=1
print("Count : ",count)
```

# Take a string input and print its reverse.

```python
s = input("Enter a string: ")
rev = ""

for ch in s:
    rev = ch + rev

print(rev)
```

# Count the number of vowels in a string.

```python
s = input("Enter a string: ")
vowel=0
for ch in s:
    if ch in ('a','e','i','o','u'):
        vowel+=1
print("Number of vowels : ",vowel)
```

# Check whether a string is a palindrome

```python
sen=input("Enter a word : ")
rev=sen[::-1]
if sen == rev:
    print("It is palindrome")
else:
    print("Its not a palindrome")
```
