```python
#Run time Polymorphism
#method overloading is not supported by python

# method overriding

class Parent:
    def show(self):
        print("This is from parent class")

class Child(Parent):
    def show(self):
        super().show()
        print("This is from child class")

obj=Child()
obj.show()
```

```python
#Encapsulation
#public variable

class Student:
    def __init__(self):
        self.name="Pratham"
    def display(self):
        print(self.name)

obj=Student()
obj.display()
```

```python
#protected variable

class Parent:
    def __init__(self):
        self._salary=80000

class Child(Parent):
    def display(self):
        print(self._salary)
obj=Child()
obj.display()
```
