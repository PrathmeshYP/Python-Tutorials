# Create a simple calculator that performs +, -, *, / based on user input.

```python
num1=int(input("Enter first number : ")) 
num2=int(input("Enter second number : "))
operator=input("Enter operator +,-,*,/")

if operator == "+":
    print("Addition = ",num1+num2)

elif operator == "-":
    print("Substraction = ",num1-num2)

elif operator == "*":
    print("Multiplication = ",num1*num2)

elif operator == "/":
    print("Division = ",num1/num2)
else :
    print("Choose correct operator")
```

# Use bitwise operators to perform AND, OR, XOR on two numbers.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Bitwise AND:", a & b)
print("Bitwise OR:", a | b)
print("Bitwise XOR:", a ^ b)
```

# Write a program to swap two numbers using bitwise XOR (without using third variable).

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Before swapping : ")
print("a = ",a)
print("b = ",b)

a=a ^ b
b=a ^ b
a=a ^ b
print("After swapping : ")
print("a = ",a)
print("b = ",b)
```
