```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("students (2).csv")
df
```

```python
#line plot

plt.plot(df['Internal_Marks'],label='Internal_Marks')
plt.plot(df['External_Marks'],label='External_Marks')
plt.title("Marks Comparison")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

```python
#bar plot
plt.bar(df['Name'],df['Internal_Marks'])
plt.title("Internal marks of student")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.show
```

```python
#histogram

plt.hist(df['Internal_Marks'],bins=10)
plt.title("Distribution of Internal Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
#scatterplot cahrt

plt.scatter(df['Internal_Marks'],df['External_Marks'])
plt.title("Internal VS External Msrks")
plt.xlabel("Internal Marks")
plt.ylabel("External Marks")
plt.show()
```

```python
# pie chart

dept_counts=df['Department'].value_counts()

plt.pie(dept_counts,labels=dept_counts.index,autopct="%1.1f%%")
plt.title("Department Distribution")
plt.show()
```

```python
#box plot

plt.boxplot(df['Project_Score'])
plt.title("Project score distribution")
plt.show()
```

```python
#subplots
plt.subplot(1,2,1)
plt.hist(df['Internal_Marks'])
plt.title("Internal Histogram")

plt.subplot(1,2,2)
plt.hist(df['External_Marks'])
plt.title("External Histogram")
```

```python
plt.bar(df['Department'],df['Semester'],color='Pink')
plt.show()
```

```python
# heat map

data=df[['Internal_Marks','External_Marks','Project_Score']]
corr=data.corr()
print(corr)
```

```python
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)),corr.columns,rotation=45)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.title("Corelation Heatmap")
plt.show()
```
