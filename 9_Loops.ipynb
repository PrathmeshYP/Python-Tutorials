# For loop

```python
for i in range(1,11):
    print(i)
```

```python
for i in range(0,11,2):
    print(i)
```

```python
fruits=["Apple","Cherry","Banana","Mango"]
for fruit in fruits:
    print(fruit)
```

```python
n=5
for i in range(1,n+1):
    print("*"*i)
    
```

```python
n=5
for i in range(n,0,-1):
    print("*"*i)
```

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
n=5
for i in range(5,0,-1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
for i in range(1,4):
    for j in range(1,4):
        if i==j:
            print("x",end=" ")
        else:
            print("o",end=" ")
    print()
```

# While loop

```python
i=1
while i<=5:
    print(i)
    i+=1
    
```

```python
for i in range(0,21):
    if i==10:
        break
    print(i)
```

```python
name="Prathmesh"
for ch in name:
    if ch=='m':
        break
    print(ch)
```

```python
for i in range(0,21):
    if i==5:
        continue
    print(i)
```
